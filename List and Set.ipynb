# What is deep copy and shallow copy.
# append,extend,concatenate difference.
# pop,remove,delete difference.

#deep and shallow copy.
dc-Creates a new object but copies references of nested objects
sc-Creates a new object and copies all nested objects completely

dc-Only outer object is copied
sc-Outer + inner (nested) objects are copied

#
append-Adds one element at the end. Modifies original list.
extend-Adds multiple elements from another iterable.modifies original list.
concatenate-Combines two lists into a new list.create a new list.

#
pop-Removes element using index.
remove-Removes element using value.
delete-Deletes element, slice, or entire list.

```python
#Rotate a list left by k positions.

num=[1,2,3,4,5]
k=2

for i in range(k):
    first=num.pop(0)
    num.append(first)
    print(num)
```

```python
#Rotate a list right by k positions.

num=[1,2,3,4,5]
k=2
for i in range(k):
    last=num.pop()
    num.insert(0,last)
    print(num)
```

```python
#Replace all even numbers with 0.

num=[1,2,3,4,5,6]

for i in range(len(num)):
    if num[i]%2==0:
        num[i]=0
print(num)
```

```python
#Separate even and odd numbers.

nums=[1,2,3,4,5,6]

even=[]
odd=[]

for num in nums:
    if num%2==0:
        even.append(num)
    else:
        odd.append(num)

print("Even:",even)
print("Odd:",odd)
```

```python
#intersection

set1={"apple","banana","cherry"}
set2={"google","microsoft","apple"}

print(set1.intersection(set2))
print(set1 & set2)
print(set1)
print(set2)
```

```python
#intersection_update()

set1={"apple","banana","cherry"}
set2={"google","microsoft","apple"}

set1.intersection_update(set2)
print(set1)
print(set2)
```

```python
#difference()

print(set1.difference(set2))
print(set1)
set3=set1-set2
print(set3)
```

```python
#difference update

set1={"apple","banana","cherry"}
set1.difference_update(set2)
print(set1)
```

```python
#symmetric Difference

set1={"apple","banana","cherry"}
set2={"google","microsoft","apple"}

set1.symmetric_difference(set2)
print(set1)
```

```python
#Symmetric_Difference_Update()

set1={"apple","banana","cherry"}
set2={"google","microsoft","apple"}

set1.symmetric_difference_update(set2)
print(set1)
```
