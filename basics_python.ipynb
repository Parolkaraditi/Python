```python
#Switch Case
#Weekdays

day=int(input("Enter day number"))

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Thursday")
    case 6:
        print("Friday")
    case 7:
        print("Saturday")
    case 8:
        print("Sunday")
    case _:
        print("Invalid day")
```

# Difference between for and while loop.

1.for loop is used to iterate over a sequence of items.
1.while loop is used to repeatedly execute a block of statements while a condition is true.

2.It requires a sequence to iterate over.
2.While the loop requires an initial condition that is tested at the beginning of the loop.

3.Number of iterations is known.
3.Number of iterations is not known.

4.Automatically handles increment/decrement.
4.we have to Iniatialize and update the variable.

5.Iterates over a sequence.
5.Repeats execution while a condition is true.

```python
# 2.

data = [1,2.5,"hi",3,"bye",True]

count_int=0
count_str=0
count_float=0
count_bool=0

for item in data:
    if type(item)==int:
        count_int+=1
    elif type(item)==str:
        count_str+=1
    elif type(item)==float:
        count_float+=1
    elif type(item)==bool:
        count_bool+=1

print("Integers:",count_int)
print("Strings:",count_str)
print("Floats:",count_float)
print("Booleans:",count_bool)
```
