```python
#Remove all duplicate characters from a string without using set().

n=input("Enter a string: ")
a=""

for ch in n:
    if ch not in a:
        a+=ch

print("String after removing duplicates:",a)
```

```python
#From a list of numbers,create a new list containing only prime numbers.

n=[10,7,3,15,11,4,13]
prime_list=[]

for num in n:
    if num>1:
        is_prime=True
        for i in range(2,num):
            if num%i==0:
                is_prime=False
                break
        if is_prime:
            prime_list.append(num)

print("Prime numbers:",prime_list)
```

```python
#Count frequency of each character in a string and store it in a dictionary.

a=input("Enter a string")
freq={}
for ch in a:
    if ch in freq:
        freq[ch]+=1
    else:
        freq[ch]=1
print("Character frequency",freq)
```

```python
#Check if two strings are anagrams.

str1=input("Enter first string:")
str2=input("Enter second string:")

str1=str1.replace("","").lower()
str2=str2.replace("","").lower()

if sorted(str1)==sorted(str2):
    print("The strings are anagrams")
else:
    print("The strings are not anagrams")


```

```python
#Find the sum of digits of a number.

num=int(input("Enter a number"))
sum=0
while num>0:
    digit=num%10
    sum += digit
    num=num//10

print("Sum of digits:",sum)
```

```python
#From a list of integers,separate positive and negative numbers into two lists.

num=[10,-5,7,-2,0,-8,15]

positive=[]
negative=[]

for n in num:
    if n>=0:
        positive.append(n)
    else:
        negative.append(n)

print("Positive numbers:",positive)
print("Negative Numbers:",negative)
```

```python
#Write a function to check whether a number is armstrong number.

n=int(input("Enter number: "))

sum=0
temp=n

while temp>0:
    digit=temp%10
    sum=sum+digit*digit*digit   
    temp=temp//10

if sum==n:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

```python
#Write a program to flatten a nested list(1 level deep).

nested=[1,2,[3,4],5,[6,7]]
flat=[]
for item in nested:
    if type(item)==list:
        flat+=item
    else:
        flat.append(item)
print(flat)



```

```python
#Merge two dictionaries.If a key exists in both,sum their values.

dict1={"a":10,"b":20,"c":30}
dict2={"b":5,"c":15,"d":40}

n=dict1.copy()
for key in dict2:
    if key in n:
        n[key]+=dict2[key]
    else:
        n[key]=dict2[key]
print(n)



```

```python
#Create a function that returns factorial using loop.

def fact(n):
    f=1
    for i in range(1,n+1):
        f*=i
    return f

num=int(input("Enter number:"))
print("Factorial:",fact(num))
```
