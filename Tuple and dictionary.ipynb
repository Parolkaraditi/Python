```python
#Tuple
#Create a tuple with 5 integers.

t=(2,3,4,5)
print(t)
```

```python
#Create a tuple with mixed data types.

my_tuple=(10,3.14,"Hello",True, [1,2,3])

print(my_tuple)
```

```python
#Access the first and last element of a tuple.

t=(10,20,30,40)

print(t[0])   
print(t[-1]) 
```

```python
#Access elements using negative indexing.

t=(10,20,30,40,50)

print(t[-1]) 
print(t[-2])  
print(t[-3]) 
```

```python
#Slice a tuple.

t=(10,20,30,40,50)

print(t[1:4])   # elements from index 1 to 3
print(t[:3])    # from start to index 2
print(t[2:]) 
```

```python
#Check the length of a tuple.

t=(10,20,30,40)

print(len(t))
```

```python
#Check if an element exists in a tuple.

t=(10,20,30,40)

print(20 in t)   
```

```python
#Create a tuple with a single element.

t=(10,)   

print(t)
print(type(t))
```

```python
# Convert a list into a tuple.

lst=[10,20,30,40]

t=tuple(lst)

print(t)
```

```python
#Convert a tuple into a list.


t=(10,20,30,40)

lst=list(t)

print(lst)
```

```python
# Concatenate two tuples.

t1=(1,2,3)
t2=(4,5,6)

t3=t1+t2

print(t3)
```

```python
#Repeat a tuple using *.

t=(1,2,3)

print(t*3)
```

```python
#Find index of an element in tuple.

t=(10,20,30,40)

print(t.index(30))
```

```python
#Count occurances of an element.

t=(10,20,30,20,40,20)

print(t.count(20))
```

```python
#Delete a tuple completely.

t=(10, 20, 30)

del t

print(t) 
```
