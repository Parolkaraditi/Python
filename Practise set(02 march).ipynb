```python
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i) + "* " * i)
```

```python
while True:
    print("\n1.Add")
    print("2.Subtract")
    print("3.Exit")

    choice=input("Enter choice")
    if choice=="1":
        a=int(input("Enter first number"))
        b=int(input("Enter second number"))
        print("Result",a+b)
    
    elif choice=="2":
        a=int(input("Enter first number"))
        b=int(input("Enter second number"))
        print("Result",a-b)

    elif choice=="3":
        print("Exiting.....")
        break
    else:
        print("Invalid Choice")
```

```python
num=int(input("Enter number"))
temp=num
power=len(str(num))
total=0

while temp>0:
    digit=temp%10
    total+=digit**power
    temp//=10

if total==num:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

```python
for num in range(1,1001):
    temp=num
    power=len(str(num))
    total=0

    while temp>0:
        digit=temp%10
        total+=digit**power
        temp//=10

    if total==num:
        print(num)
```

```python
list=[5,3,8,1,2]
for i in range(len(list)):
    for j in range(len(list)-1):
        if list[j]>list[j+1]:
            list[j],list[j+1]=list[j+1],list[j]
print("Sorted List",list)
```
