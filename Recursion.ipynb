```python
#Recursion=when a function calls itself to solve a problem.

def countdown(n):
    if n <= 0:
        print("Done!")
    else:
        print(n)
        countdown(n - 1)

countdown(5)
```

```python
#Factorial

def fact(n):
    if n<=1:
        return 1
    return n*fact(n-1)

num=int(input("Enter Number:"))
print("Factorial=",fact(num))
    

```

```python
#Sum

def sum(n):
    if n<=1:
        return n
    return n+sum(n-1)
num=int(input("Enter any number"))
print("Sum=",sum(num))

```

```python
#Print numbers 1-5.

def print_numbers(n):
    if n==0:
        return 0
    print_numbers(n-1)
    print(n)
print_numbers(5)
```
